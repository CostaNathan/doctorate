```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
header = pd.MultiIndex.from_product([['EPC VILA ZILDA - VAZÕES - 2011 A 2020', 'EPC SANTOS - VAZÕES - 2011 A 2020', 'EPC CAIÇARA - VAZÕES - 2011 A 2020', 'EMISSÁRIO FORTE - VAZÕES - 2011 A 2020', 'EMISSÁRIO TUPI- VAZÕES - 2011 A 2020'], # Multiindex 
                                     ['Vazão (L/s)', 'vazao_maxima','vazao_minima']], 
                                    names = ['Emissários', 'Vazão'])


vazao_dummy = pd.read_excel('https://github.com/CostaNathan/doctorate/blob/b42b9dbb2180caad65804ea413a08796636bffdd/data/Monitoring_points/vazoes_corrigidas.xlsx?raw=true')
vazao_dummy = vazao_dummy.replace(['-'], np.nan)

index = []
datestring = vazao_dummy['Unnamed: 0'].astype(str)
for i in range (1,len(datestring)):
  index.append(datetime.strptime(datestring[i], "%Y-%m-%d %H:%M:%S")) # Datetime set up

# vazao_dummy
# datestring
# index

vazao_discharge = pd.DataFrame(vazao_dummy.iloc[1:,1:16].values, index = index, columns= header, dtype=float) # Parameter dataframe
vazao_discharge
```

```python
# Inserção dos dias correspondentes aos meses para usar no Delft3D

def dias_mes_normal():
  day = []
  for i in range(12):
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 1:
      day.append(0)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 2:
      day.append(0+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 3:
      day.append(31+28)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 4:
      day.append(31+28+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 5:
      day.append(31+28+31+30)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 6:
      day.append(31+28+31+30+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 7:
      day.append(31+28+31+30+31+30)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 8:
      day.append(31+28+31+30+31+30+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 9:
      day.append(31+28+31+30+31+30+31+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 10:
      day.append(31+28+31+30+31+30+31+31+30)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 11:
      day.append(31+28+31+30+31+30+31+31+30+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 12:
      day.append(31+28+31+30+31+30+31+31+30+31+30)
    
  return day

def dias_mes_bissexto():
  day = []
  for i in range(12):
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 1:
      day.append(0)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 2:
      day.append(0+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 3:
      day.append(31+29)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 4:
      day.append(31+29+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 5:
      day.append(31+29+31+30)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 6:
      day.append(31+29+31+30+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 7:
      day.append(31+29+31+30+31+30)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 8:
      day.append(31+29+31+30+31+30+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 9:
      day.append(31+29+31+30+31+30+31+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 10:
      day.append(31+29+31+30+31+30+31+31+30)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 11:
      day.append(31+29+31+30+31+30+31+31+30+31)
    if vazao_discharge.index.month[vazao_discharge.index.year == 2011][i] == 12:
      day.append(31+29+31+30+31+30+31+31+30+31+30)
  return day
# print(dias_mes_normal(), dias_mes_bissexto())

dias = [dias_mes_normal(),dias_mes_bissexto(),dias_mes_normal(),dias_mes_normal(),dias_mes_normal(),dias_mes_bissexto(),dias_mes_normal(),dias_mes_normal(),dias_mes_normal(),dias_mes_bissexto()]
dias = np.array(dias) 
flattendias = dias.flatten()
# flattendias

vazao_discharge['dia'] = flattendias
vazao_discharge
```

```python
# Inserção das horas,minutos e segundos correspondentes aos meses para usar no Delft3D

hora = np.zeros(len(vazao_discharge.index.tolist()))
minutos = np.zeros(len(vazao_discharge.index.tolist()))
segundos = np.zeros(len(vazao_discharge.index.tolist()))

hora = [int(x) for x in hora]
minutos = [int(x) for x in minutos]
segundos = [int(x) for x in segundos]

vazao_discharge['hora'] = hora
vazao_discharge['minutos'] = minutos
vazao_discharge['segundos'] = segundos

vazao_discharge
```

#### Vazão Santos estrutura Delft3D

```python
vazao_santos = pd.concat([vazao_discharge.loc[:,('dia')],vazao_discharge.loc[:,('minutos')],vazao_discharge.loc[:,('segundos')],vazao_discharge.loc[:,('EPC VILA ZILDA - VAZÕES - 2011 A 2020')]],axis = 1, levels = 0)

vazao_santos
```
