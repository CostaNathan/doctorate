```python
inmet_CWB = pd.read_csv('dados_A807_H_2019-12-01_2020-08-01_OK.csv',
                            sep=';', skiprows=10)
inmet_CWB
```

```python
inmet_CWB['RADIACAO GLOBAL(W/m²)']
```

```python
datamedicao = inmet_CWB['Data Medicao'] # Retorna uma série
print(datamedicao, type(datamedicao))
```

```python
datamedicao = inmet_CWB['Data Medicao'].values # Retorna um array
horamedicao = inmet_CWB['Hora Medicao'].values

print(datamedicao, '\n', horamedicao, '\n', type(datamedicao))
```

```python
datahorario = []

for i in range(len(datamedicao)):
  datahorario.append(datamedicao[i] + ' ' + str(int(horamedicao[i] // 100))) 
  
  # // operator = floor division

datahorario[0:2] # Retorna uma lista com a data + hora 
```

```python
# datetime.strptime

date_string = "21 June, 2018"

print("date_string =", date_string)
print("type of date_string =", type(date_string))

date_object = datetime.strptime(date_string, "%d %B, %Y")

print("date_object =", date_object)
print("type of date_object =", type(date_object))

#    %d - Represents the day of the month. Example: 01, 02, ..., 31
#    %B - Month's name in full. Example: January, February etc.
#    %Y - Year in four digits. Example: 2018, 2019 etc.

dt_string = "12/11/2018 09:15:32"

# Considering date is in dd/mm/yyyy format
dt_object1 = datetime.strptime(dt_string, "%d/%m/%Y %H:%M:%S")
print("dt_object1 =", dt_object1)

# Considering date is in mm/dd/yyyy format
dt_object2 = datetime.strptime(dt_string, "%m/%d/%Y %H:%M:%S")
print("dt_object2 =", dt_object2)
```

```python
lista = []

for i in range(len(datamedicao)):
  lista.append(datetime.strptime(datahorario[i], '%Y-%m-%d %H')) 
  # The strptime() method creates a datetime object from the given string

lista[:24]
```

```python
inmet_CWB['data'] = lista # Cria uma nova columa chamada data com os valores da variavel lista
inmet_CWB
```

```python
inmet_CWB = inmet_CWB.set_index('data')
inmet_CWB
```

```python
del inmet_CWB['Data Medicao']
del inmet_CWB['Hora Medicao']

inmet_CWB
```

```python
inmet_CWB_diario = inmet_CWB.resample('d').mean() # d = dia

inmet_CWB_diario # Contem somente os dias e médias
```
