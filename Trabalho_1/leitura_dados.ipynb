```python
lista = []
for i in range(1, 13):
  var = dados_barragem[dados_barragem['Mês'] == i]['Vel. Vento média (m/s)']
  lista.append(round(var.mean(), 2))

maior = len(str(lista[0]))
for l in lista:
  if len(str(l)) > maior:
    maior = len(str(l))

cont = 1
for l in lista:
  print(f'{l} \t{" " * maior if len(str(l)) < maior - 1 else ""}{cont}')
  cont += 1
```
