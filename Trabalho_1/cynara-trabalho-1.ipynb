```python
# Para calcular uma variável variável dentro do odeint

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


def k(t):
    if t >= 5:
        return (t - 20) + 2 * t
    else:
        return 1
    
def β(t):
    global H
    
    H = 10

    if H <= 5:
        return (H - 20) + 2 * t
    else:
        return 2


def model(y, t, k, β):
    
    dydt = -k(t) * y + β(t) * H

    return dydt


y0 = 5
t = np.linspace(0, 10)
args = (k, β)
y = odeint(model, y0, t, args)

fig = plt.figure(dpi=100)
ax = fig.add_subplot(111)
(h1, ) = ax.plot(t, y)
(h2, ) = ax.plot(t, [k(s) for s in t])
(h3, ) = ax.plot(t, [β(s) for s in t])
ax.legend([h1, h2, h3], ["y", "a", "β"])
ax.set_xlabel("t")
ax.grid()
plt.show()
```

```python
# Modelagem nitrificação (CHAPRA, p. 426)
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

T = temperatura_trib_F5

c = 1
def contador():
    global c
    
    c += 1
    
    return c


def kd(t):
    global H

    H = 3.4 # chute

    if 0 <= H <= 2.4:
        return 0.3 * (H / 2.4) ** (- 0.434)
    elif H > 2.4:
        return 0.3


def kdz(t):
    global T
    
    kdz_20 = 1 # chute
    theta = 1 # chute
    
#     contador()

#     return kdz_20 * theta ** (T[c] - 20)
#     return kdz_20 * theta ** (np.mean(T) - 20)
    return kdz_20 * theta ** (T[c] if c <= len(T) - 1 else T[len(T) - 1] - 20)


def odes(x, t, kd, kdz):
    
    # Variáveis
    V = 1 # chute
    os = 8 # chute
    ka = 0.1 # chute
    kra = 1 # chute
    krz = 1 # chute
    kn = 1 # chute
    ksed = 0.02 # chute
    roc = 2.6667 # gOD/gC
    SOD = 1 # chute
    rca = 1 # chute
    pa = 1 # chute
    alpha = 0.01
    LN = 1 # chute
    L0 = 10 # chute
    Lpart = 1 # chute
    Czooplancton = 1 # chute
    roa = 0 # chute
    roi = 0 # chute
    ron = roa + roi
    z = 1 # chute

    o = x[0]
    L = x[1]
    
    # Equações OD:
    dodt_termo_1 = ka * V * (os - o)
    dodt_termo_2 = - kd(t) * V * L0 * np.exp(- kd(t) * t)
    dodt_termo_3 = - ron * V * kn * LN
    dodt_termo_4 = pa * V
    dodt_termo_5 = - rca * roc * V * kra * alpha
    dodt_termo_6 = roc * V * krz * Czooplancton
    dodt_termo_7 = V * (- SOD / H)

    dodt = (dodt_termo_1 + dodt_termo_2 + dodt_termo_3 + dodt_termo_4 + dodt_termo_5 + dodt_termo_7) / V

    # Equações DBO:
    dLdt_termo_1 = - kd(t) * V * L
    dLdt_termo_2 = - ksed * (1 - Lpart) * L
    dLdt_termo_3 = rca * roc * V * kra * alpha
    dLdt_termo_4 = roc * V * kdz(t) * z

    dLdt = dLdt_termo_1 + dLdt_termo_2 + dLdt_termo_3 + dLdt_termo_4

    # An oxygen deficit balance can be written as:
    # dDdt = roa * kai * Na + roi * kin * Ni - ka * D
    
    contador() # para mudar o contador

    return dodt, dLdt

o0 = 0; L0 = 0

x0 = [o0, L0]

t = np.linspace(0, len(T), 100)

args = (kd, kdz)

x = odeint(odes, x0, t, args)

x_1 = x[:, 0]
x_2 = x[:, 1]

plt.plot(t, x_1, marker='o', label='OD')
plt.plot(t, x_2, marker='o', label='DBO')
plt.legend()
plt.show()

print(len(x_1), len(x_2), len(T))
```
