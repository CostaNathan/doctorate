#### Reaeração (Chapra, 1997)
Como a reaeração possui elevada constante de Henry (aprox. 0.8 atm $m^{3} mole^{-1}$) o processo é governado pela fase líquida do sistema. Sendo assim a equação geral do fluxo para qualquer gás é modificada:

de
\begin{equation}
J = \nu_{\nu}(\frac{p_{g}}{H_{e}}-c_{l})
\end{equation}
para
\begin{equation}
J = K_{l}(o_{s}-o)
\end{equation}
Onde $K_{l}$ é o coeficiente de transferência de massa de oxigênio; $o_{s}$ é a concentração de oxigênio de saturação; $o$ é a concentração de oxigênio medido.

Por fim foi convertido o fluxo de moles para um fluxo de massa por tempo a partir da multiplicação dos dois lados da equação acima pelo peso do oxigênio (32g/mol) e pela área superficial de líquido exposta a atmosfera, resultando em:

\begin{equation}
V\frac{do}{dt} = K_{l}A_{s}(o_{s}-o)
\end{equation}
Sendo que
\begin{equation}
V=A_{s}H
\end{equation}
Logo
\begin{equation}
V\frac{do}{dt} = k_{a}V(o_{s}-o)
\end{equation}

Onde $k_{a}$ é a taxa de reaeração e equivale a:

\begin{equation}
k_{a} = \frac{K_{l}}{H}
\end{equation}

e variável com a temperatura conforme a expressão a seguir:

\begin{equation}
k_{a,T} = k_{a,20}θ^{T-20}
\end{equation}

Onde θ ≈ 1.024

A constante de reaeração também pode ser adotada conforme alguns experimentos empíricos, sendo eles algumas das fórmulas mais usadas O'Connor-Dobbins, Churchill, e Owens-Gibbs. Segundo Chapra (1997), o primeiro possui a maior aplicabilidade entre os três, enquanto o segundo para corpos d'água mais rápidos e a última é utilizada pra sistemas rasos, assim como mostra a Figura abaixo.

Deste modo foi adotado a constante de reaeração conforme O'Connor-Dobbins, que pode ser expressa como:

\begin{equation}
k_{a} = 3.93\frac{U^{0.5}}{H^{1.5}}
\end{equation}

Com $k_{a}$($d^{-1})$, U (mps), H (m)




### Modelo de Streeter-Phelps

Esse modelo amarra os doi principais mecanismos governando tanto a variação de oxigenio dissolvido em um corpo d'água sob influência de uma fonte lançando efluente, quanto a decomposição da materia organica e o oxigênio resultante da aeração.

A partir dele obtêm-se as equações governantes acopladas entre a variação de DBO e OD:

\begin{equation}
V\frac{dL}{dt} = -k_{d}VL
\end{equation}
e 

\begin{equation}
V\frac{dD}{dt}=k_{d}VL - k_{a}VD
\end{equation}

onde:

L = concentração de DBO em mg/L

D = $o_{s} - o$ = Deficit de oxigênio dissolvido

$k_{a}=$ Constante de reaeração $(d^{-1})$

$k_{d}=$ Constante de fluxo mássico de DBO (carbonáceo)

Podendo ser expresso como:

$k_{d}=0.3(\frac{H}{2.4})^{-0.434}$ para $0 \leq H \leq 2.4m $

$k_{d}=0.3$ para $H > 2.4m $

##### Decomposição
A representação da DBO ocorre a partir da utilização da nova variável $ L (mgO L^{-1})$ que representa a quantidade de matéria organica oxidável restante na garrafa (garrafa DBO):

\begin{equation}
V\frac{d[DBO]}{dt} = V\frac{dL}{dt} = -k_{1}VL
\end{equation}
Se $L=L_{0}$ para t=0
\begin{equation}
L(t) = L_{0} e^{-k_{1}*t}
\end{equation}

O oxigênio consumido durante a decomposição pode ser definido como:
\begin{equation}
y=L_{0} - L = L_{0}(1-e^{-K_{1}t}) 
\end{equation}
com $y = BOD (mgO L^{-1})$.
