### Modelo DBO e $O_{2}$ 

Modelo utilizando como base um sistema em batelada fechado

#### Balanço DBO
A representação da DBO ocorre a partir da utilização da nova variável $ L (mgO L^{-1})$ que representa a quantidade de matéria organica oxidável restante na garrafa (garrafa DBO):

\begin{equation}
V\frac{d[DBO]}{dt} = V\frac{dL}{dt} = -k_{1}VL
\end{equation}
Se $L=L_{0}$ para t=0
\begin{equation}
L(t) = L_{0} e^{-k_{1}*t}
\end{equation}

O oxigênio consumido durante a decomposição pode ser definido como:
\begin{equation}
y=L_{0} - L = L_{0}(1-e^{-K_{1}t}) 
\end{equation}
com $y = BOD (mgO L^{-1})$.

#### Balanço $O_{2}$

\begin{equation}
V\frac{dO}{dt} = -k_{1}VL_{0}e^{-k_{1}t}
\end{equation}

Se $ O=O_{0}$ para t=0
\begin{equation}
O = O_{0} - L_{0}(1-e^{-k_{1}t})
\end{equation}


O consumo do $O_{2}$ também é afetado pelo processo de nitrificação, deste modo requerendo um sistema acoplado considerando tanto a DBO quanto a NDBO (demanda de oxigênio devido a nitrificação)


Assumindo um córrego em estado estacionário (steady-state = $\frac{d}{dt}=0$), com mistura instantânea (sem variação espacial da substância) e adotando $kr= k_{d} + k_{s}$, onde $kd =$ taxa de decomposição no rio $(d^{-1})$ e $k_{s} = \frac{\nu_{s}}{H}=$ taxa de sedimentação  $(d^{-1})$:

\begin{equation}
\frac{dL}{dt} = -U\frac{dL}{dx} - k_{r}L
\end{equation}
Estado estacionário:
\begin{equation}
0 = -U\frac{dL}{dx} - k_{r}L
\end{equation}
Mistura instantânea:
\begin{equation}
L_{0} = \frac{Q_{w}L_{w} + Q_{r}Q_{r}}{Q_{w}+Q_{r}}
\end{equation}
Logo: 
\begin{equation}
L = L_{0}e^{-\frac{k_{r}}{U^{x}}}
\end{equation}
