\begin{equation}
J = \nu_{\nu}(\frac{p_{g}}{H_{e}}-c_{l})
\end{equation}


\begin{equation}
J = K_{l}(o_{s}-o)
\end{equation}
Onde $K_{l}$ é a constante de reaeração; $o_{s}$ é a concentração de oxigênio de saturação; $o$ é a concentração de oxigênio medido

##### Decomposição
A representação da DBO ocorre a partir da utilização da nova variável $ L (mgO L^{-1})$ que representa a quantidade de matéria organica oxidável restante na garrafa (garrafa DBO):

\begin{equation}
V\frac{d[DBO]}{dt} = V\frac{dL}{dt} = -k_{1}VL
\end{equation}
Se $L=L_{0}$ para t=0
\begin{equation}
L(t) = L_{0} e^{-k_{1}*t}
\end{equation}

O oxigênio consumido durante a decomposição pode ser definido como:
\begin{equation}
y=L_{0} - L = L_{0}(1-e^{-K_{1}t}) 
\end{equation}
com $y = BOD (mgO L^{-1})$.
